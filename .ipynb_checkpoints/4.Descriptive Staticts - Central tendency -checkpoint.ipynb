```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
diamonds = sns.load_dataset("diamonds")
diamonds.describe()
```

# Central tendency

```python
# identify the center or typical value around which the data tend to cluster.
```

# Mean

```python
#The Mean is the sum of all values divided by the number of values.

info=diamonds.describe()              #mean of all quantitave data in row index 1
mean = info.iloc[1]
for index,value in mean.items():
    print(f"mean of {index} is {value:.3f}")
```

```python
#when the data have significantly different points from the majority of the rest values (Outliers) -> Mean can't represent center of the data
# that is what's called (SKEWNESS) -> will be mentioned later
```

# Median

```python
# The Median is the middle value of a dataset when the data is sorted in ascending or descending order .
# More representative when data have outliers
info=diamonds.describe()              #Median of all quantitave data in row index 5
median = info.iloc[5]
for index,value in median.items():
    print(f"median of {index} is {value:.3f}")
```

# Mode 

```python
# The Mode is the value that occurs most frequently in a dataset.
# most popular with categorical data
mode=diamonds["color"].mode()
print("Mode of color:", mode[0])
```

# Skewness

```python
# Skewness measures the asymmetry of a data distribution around its mean.
# larger values(skew) pull the mean towards it.
```

### Right Skweness

```python
# Distribution skews to the right => Long tail on the right (larger values) ..
# Mean > Median > Mode

plt.figure(figsize=(10,7))
plt.xlim((0, diamonds["price"].max()))
plt.title("Right skewness with Central tendency measures", fontsize=14)

price_mu=diamonds['price'].mean()
price_median=diamonds['price'].median()
price_mode =diamonds['price'].mode()[0]

plt.vlines(price_mu,colors='r',lw=2,ymin=0,ymax=7000, label=f"Mean ({price_mu:.2f})")
plt.vlines(price_median,colors='g',lw=2,ymin=0,ymax=7000, label=f"Median ({price_median:.2f})")
plt.vlines(price_mode,colors='k',lw=2,ymin=0,ymax=7000, label=f"Mode ({price_mode:.2f})")

sns.histplot(diamonds['price'],kde=True)
plt.legend(title="Central Tendency", loc="upper right", frameon=True, fancybox=True, shadow=True)
plt.xlabel("Price")
plt.ylabel("Frequency")

diamonds['price'].skew()
```

### Zero Skweness

```python
# Distribution is symmetric around the center (Often a Normal distribution) .
# Mean = Median = Mode

plt.figure(figsize=(10,7))
plt.xlim((30, diamonds["depth"].max()))
plt.title("Zero skewness with Central tendency measures", fontsize=14)

depth_mu=diamonds['depth'].mean()
depth_median=diamonds['depth'].median()
depth_mode =diamonds['depth'].mode()[0]

plt.vlines(depth_mu,colors='r',lw=2,ymin=0,ymax=2000,label=f"Mean ({depth_mu:.2f})")
plt.vlines(depth_median,colors='g',lw=2,ymin=0,ymax=1500, label=f"Median ({depth_median:.2f})")
plt.vlines(depth_mode,colors='k',lw=2,ymin=0,ymax=1750, label=f"Mode ({depth_mode:.2f})")

sns.histplot(diamonds['depth'],kde=True)

plt.legend(title="Central Tendency", loc="upper right", frameon=True, fancybox=True, shadow=True)
plt.xlabel("depth")
plt.ylabel("Frequency")

diamonds['depth'].skew()
```

### Left Skweness

```python
# Distribution skews to the left => Long tail on the left (larger values) ..
# Mean < Median < Mode
```

```python
# Left skewed Data generation using Beta distribution
data_left = np.random.beta(a=5, b=2, size=1000) * 100  # scale to 0-100
df = pd.DataFrame(data_left, columns=['Values'])

plt.figure(figsize=(10,7))
plt.title("Left Skewed Distribution with Central Tendency Measures", fontsize=14)

df_mu = df['Values'].mean()
df_median = df['Values'].median()
df_mode = df['Values'].mode()[0]

sns.histplot(df['Values'], kde=True, color='skyblue')
plt.vlines(df_mu, ymin=0, ymax=200, colors='r', lw=2, label=f"Mean ({df_mu:.2f})")
plt.vlines(df_median, ymin=0, ymax=200, colors='g', lw=2, label=f"Median ({df_median:.2f})")
plt.vlines(df_mode, ymin=0, ymax=200, colors='k', lw=2, label=f"Mode ({df_mode:.2f})")

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend(title="Central Tendency", loc="upper right", frameon=True, fancybox=True, shadow=True)
plt.show()

df['Values'].skew()
```

### how to deal with skewness 

```python
# Use transformations that correct extreme Data
# ex => right skewness -> Log transform , Root , Box-Cox

# applying Log to the diamond price
diamonds["log_price"] = diamonds["price"].apply(np.log)

plt.figure(figsize=(10,7))
plt.xlim(( diamonds["log_price"].min(), diamonds["log_price"].max()))
plt.title("Right skewness with Central tendency measures", fontsize=14)

log_price_mu=diamonds["log_price"].mean()
log_price_median=diamonds["log_price"].median()
log_price_mode =diamonds["log_price"].mode()[0]

plt.vlines(log_price_mu,colors='r',lw=2,ymin=0,ymax=7000, label=f"Mean ({log_price_mu:.2f})")
plt.vlines(log_price_median,colors='g',lw=2,ymin=0,ymax=7000, label=f"Median ({log_price_median:.2f})")
plt.vlines(log_price_mode,colors='k',lw=2,ymin=0,ymax=7000, label=f"Mode ({log_price_mode:.2f})")

sns.histplot(diamonds["log_price"],kde=True)
plt.legend(title="Central Tendency", loc="upper right", frameon=True, fancybox=True, shadow=True)
plt.xlabel("Log Price")
plt.ylabel("Frequency")
```
