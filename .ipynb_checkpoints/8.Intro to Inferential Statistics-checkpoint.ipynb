```python
import numpy as np
from scipy.stats import expon,skew,norm,t,sem
import seaborn as sns
import matplotlib.pyplot as plt
```

# Sampling Distribution

```python
# the probability distribution of a statistic (like a mean, variance, etc.) that you get from repeated random samples of the same population.
# EX:Imagine you have a population (like all people in a country) , and U take many random samples => if you get the mean for all samples ,the distribution of all means is a sampling distribution
```

# central limit theorem (CLT)

```python
# Even if your population is not normally distributed, the mean distribution of many random samples n will be approximately normal .
# EX :
mean = 5
population = expon.rvs( loc = 0 , scale = mean , size = 10000 )         # exponential distribuion (Not normal)
sns.histplot(x=population,stat='count')
```

```python
    # taking 500 samples of n numbers out of the populaion
    plt.figure(figsize=(20,3))
    n = [3 , 100 , 500 ]
    for index,size in enumerate(n):
        plt.subplot(1,3,index+1)
        samples = [np.random.choice(population, size = size, replace=False) for _ in range(1000)]
        samples_mean = [np.mean(samples[i]) for i in range(1000)]             # mean of all samples (sampling distribution)
        plt.title(f"""Sampling distribution of sample size = {size}
        mean= {np.mean(samples_mean):.2f} & variance = {np.var(samples_mean):.2f}
        skewness = {skew(samples_mean):.2f} """)
        ax= sns.histplot(x=samples_mean,stat='count')                                   # approximately normal and tends to symmetry increasing n 
```

### Conclusion

```python
# whatever the population disribution looks like ..it's sampling distribution looks normal 
# the mean of the sampling mean = the mean of the population
# using bigger sample size norrows the sampling distribution (var << )
# vaiance of the sampling distribution = variance of the population / n
```

# Estimasion Types 

## 1. Point Estimation (best guess)

```python
# Point estimation gives a single numerical value (a point) as an estimate of an unknown population parameter (like the mean or variance).
# ex : Mean of the population is m
```

## 2. Interval Estimation

```python
# Interval estimation gives a range of values (an interval) that is likely to contain the true population parameter, instead of just one point.
# Example : I am 95% confident that the true mean lies between 80 and 84.
#--------------------------------Confidence Interval (CI)I------------------------------------
# CL = P( -C < statistic < +C )
# CI = X  Error margin of the CL
# Error margin = C * (std / n)
# C -> critical value (Z or T) depending on the Data
```

### Z distribution VS T distribution


```python
#--------------------------------Z distribution-------------------------------------------- 
# uses the actual std of the population () which is unknown and can't be determined easily
# underestimate the error margin (it assume u know )   
# narrower -> more certain 
# used when dealing with the population or large sample of it


#-------------------------------- T distribution-------------------------------------------- 
# uses the approximate std calculated out of the sample (S)
# As df increases(>1000)  T distribution approaches Z distribution
# wider -> less certain
# used when dealing with small sample n< 1000

# ------------------------------------------------------------------------------------------
x = np.linspace(-5,5,1000)
Z = norm.pdf(x,0,1)
T3 = t.pdf(x,3,0,1)
T10 = t.pdf(x,10,0,1)
T1000 = t.pdf(x,1000,0,1)
plt.figure(figsize=(10,6))
plt.plot(x, Z , label='Z Distribution ( known)', color='black', linewidth=2)
plt.plot(x, T3 , label='t Distribution (df=3)', linestyle='--')
plt.plot(x, T10 , label='t Distribution (df=10)', linestyle='--')
plt.plot(x, T1000 , label='t Distribution (df=1000)', linestyle='--')
plt.title("Z vs t Distribution", fontsize=16)
# approch to Z distribution by increasing n


#---------------------------------- Critical Values ----------------------------------
Z_value = norm.ppf(0.5+0.95/2)
T3_value = t.ppf(0.5+0.95/2,df=3)
T10_value = t.ppf(0.5+0.95/2,df=10)
T1000_value = t.ppf(0.5+0.95/2,df=1000)

print (f"""Z value = {Z_value}
T value at df:3 = {T3_value}
T value at df:10 = {T10_value}
T value at df:1000 = {T1000_value}
""")

plt.axvline(Z_value, color='black', linestyle='-', linewidth=2, label=f'Z = {Z_value:.2f}')
plt.axvline(-Z_value, color='black', linestyle='-', linewidth=2)
plt.axvline(T3_value, color='blue', linestyle=':', label=f'T(df=3) = {T3_value:.2f}')
plt.axvline(-T3_value, color='blue', linestyle=':')
plt.axvline(T10_value, color='orange', linestyle=':', label=f'T(df=10) = {T10_value:.2f}')
plt.axvline(-T10_value, color='orange', linestyle=':')
plt.axvline(T1000_value, color='green', linestyle=':', label=f'T(df=1000) = {T1000_value:.2f}')
plt.axvline(-T1000_value, color='green', linestyle=':')

# nearly T = Z at n = 1000

plt.legend()
plt.grid(False)
plt.show()
```

```python
#----------------------------------Example : Confidence Interval at CL = 95 % ----------------------------------
sample = np.random.normal(loc=50, scale=5, size=10)
population = np.random.normal(loc=50, scale=5, size=1000)

sample_n = len(sample)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)   # ddof=1 => sample std

population_n = len(population)
population_mean = np.mean(population)
population_std = np.std(population)

z_crit = norm.ppf(0.5 + 0.95/2 )
t_crit = t.ppf(0.5 + 0.95/2, df=sample_n-1)

population_margin_error = z_crit * (population_std / np.sqrt(population_n))
sample_margin_error = t_crit * (sample_std / np.sqrt(sample_n))

CI_population = [population_mean - population_margin_error,
            population_mean + population_margin_error]

CI_sample = [sample_mean - sample_margin_error,
            sample_mean + sample_margin_error]



print(f""" confindence interval of the sample is {CI_sample}

 confindence interval of the population is {CI_population}
""")
```

```python
t.interval(0.95,
           df=sample_n-1,
           loc=sample_mean,
           scale=sem(sample)
          )
```

```python
t.interval(0.95,
           df=population_n-1,
           loc=population_mean,
           scale=sem(population)
          )
```
