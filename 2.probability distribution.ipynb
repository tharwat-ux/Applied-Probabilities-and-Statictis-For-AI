```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint , norm
```

## probability distribution

```python
# A probability distribution shows how probabilities are assigned to all possible outcomes of a random variable.
```

### Types

```python
# 1- Discrete Probability Distributions : Used when the random variable takes finite values.
# PMF gives the exact probability of each possible value.
    #example : roll die
discrete_sample_space = list(range(1, 7))
start = 1
end = 6
pmf = randint.pmf(discrete_sample_space,start,end+1)
plt.stem(discrete_sample_space, pmf, linefmt='-', markerfmt='o', basefmt=" ")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.ylim(0, max(pmf) + 0.01)
plt.title("PMF of a Fair Dice Roll")
plt.xticks(discrete_sample_space)
plt.show()
```

```python
#2- Continuous Probability Distributions : Used when the random variable takes infinite values.
#PDF gives the density of probability — not the exact probability.
    #example : grades 
continuos_sample_space = np.linspace(0,100,1000)
pdf = norm.pdf(continuos_sample_space,loc=np.mean(continuos_sample_space),scale=10)

# to calculate the probability you should calc area under the curve (integration) or use CDF
continuos_sample_space = np.linspace(0,100,1000)
CDF = norm.cdf(continuos_sample_space,loc=np.mean(continuos_sample_space),scale=10) 


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))  # 1 صف, 2 عمود

ax1.plot(continuos_sample_space, pdf, color='blue')
ax1.set_title('PDF (Density)')
ax1.set_xlabel('outcomes')
ax1.set_ylabel('Density')
ax1.grid(True)

ax2.plot(continuos_sample_space, CDF, color='green')
ax2.set_title('CDF (Cumulative Probability)')
ax2.set_xlabel('outcomes')
ax2.set_ylabel('Probability')
ax2.grid(True)

plt.tight_layout()
plt.show()
```
